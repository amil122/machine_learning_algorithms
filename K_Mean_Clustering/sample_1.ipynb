```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
num_records = 100  # Number of records you want in your dataset
salaries = np.random.randint(30000, 100000, size=num_records)  # Salaries between 30k and 100k
expenditures = np.random.randint(5000, 30000, size=num_records)  # Expenditures between 5k and 30k
```

```python
df = pd.DataFrame({
    'Salary': salaries,
    'Expenditure': expenditures
})

# Display the first few rows of the DataFrame
print(df.head())
```

```python
sns.scatterplot(data = df, x= df['Salary'], y = df['Expenditure'])
```

```python
scale = MinMaxScaler()
scaled_df = scale.fit_transform(df)
```

```python
dff = pd.DataFrame(data = scaled_df ,columns=df.columns)
```

```python
dff
```

```python
individual_clustering_score = []
for i in range(1,9):
    kmeans = KMeans(n_clusters = i, init='k-means++')
    kmeans.fit(dff)
    individual_clustering_score.append(kmeans.inertia_)
```

```python
plt.plot(range(1,9),individual_clustering_score)
plt.xlabel('No. of cluster')
plt.ylabel('clustering score')
```

```python
kmeans = KMeans(n_clusters = 2, init='k-means++' )
kmeans.fit(dff)
pred = kmeans.predict(dff)
pred
```

```python
dff['Cluster'] = pd.DataFrame(pred)
```

```python
dff
centers = kmeans.cluster_centers_
centers_df = pd.DataFrame(data=centers, columns=['Feature 1', 'Feature 2'])
centers_df
```

```python
sns.scatterplot(data =dff, x ='Salary',y ='Expenditure', hue = 'Cluster')
sns.scatterplot(data=centers_df, x='Feature 1', y='Feature 2', marker='^', s=100, color='red', label='Cluster Centers')
```
