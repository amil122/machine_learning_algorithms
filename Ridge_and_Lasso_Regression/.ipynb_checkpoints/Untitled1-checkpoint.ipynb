```python
columns_fill_zero = ['Bedroom2','Bathroom','Landsize','Propertycount','Distance','Postcode']
```

```python
df[columns_fill_zero] = df[columns_fill_zero].fillna(0)
```

```python
mean_building_area = df['BuildingArea'].mean()
df['BuildingArea'] = df['BuildingArea'].fillna(mean_building_area)
```

```python
mean_Lattitude = df['Lattitude'].mean()
df['Lattitude'] = df['Lattitude'].fillna(mean_Lattitude)
```

```python
mean_Longtitude = df['Longtitude'].mean()
df['Longtitude'] = df['Longtitude'].fillna(mean_Longtitude)
```

```python
df.isna().sum()
```

```python
mean_yr = df['YearBuilt'].mean()
df['YearBuilt'] = df['YearBuilt'].fillna(mean_yr)
```

```python
df.dropna(inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Type'] = encoder.fit_transform(df['Type'])
```

```python
object_cols = df.select_dtypes(include = ['object'])
dataset = pd.get_dummies(object_cols, drop_first = False)
```

```python
df_combined = pd.concat([df, dataset], axis=1)
df_combined
```

```python
df_combined.info()
```

```python
obj_cols = df_combined.select_dtypes(include=['object']).columns

# Step 2: Drop object-type columns from the DataFrame
df_combined.drop(columns=obj_cols, inplace=True)
```

```python
df_combined
```

```python
x = df_combined.drop('Price', axis = 1)
y = df_combined['Price']
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30)
```

```python
print(x.shape)  # (num_samples, num_features)
print(y.shape)  # (num_samples,)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
score = reg.score(x_test,y_test)
print(score)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
estimator = Lasso()
params = {'alpha': [1e-15,1e-10,1e-8,1e-2,1,45,50,55,60,64,70,75,100]}
lasso_reg = GridSearchCV(estimator, params,cv = 4)
```

```python
lasso_reg.fit(x_train,y_train)
```

```python
print(lasso_reg.best_params_)
print(lasso_reg.best_score_)
```
